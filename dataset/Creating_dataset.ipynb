```python
def join_data(data):
  return ('\n').join(data)
```

```python
def split_data(data):
  new_text = []
  step = 12
  limit = 600

  for i in range(0, len(data)):
    text = data['content'][i]
    

    if '[' in text:
      text = text.replace('[', '')
    if ']' in text:
      text = text.replace(']', '')
    if '<' in text:
      text = text.replace('<', '')
    if '>' in text:
      text = text.replace('>', '')
    if '\xa0' in text:
      text = text.replace('\xa0', '')

    text = text.split('\n')
    

    for i in range(0, len(text), step):
      if len(('\n').join(text[i:i+step])) > limit:
        print(len(new_text))
        print(('\n').join(text[i:i+step]))
        continue
      new_text.append(('\n').join(text[i:i+step]))

  return new_text
```
